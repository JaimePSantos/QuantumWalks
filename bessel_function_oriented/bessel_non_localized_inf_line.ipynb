```python
import numpy as np
import scipy.special as sp
import matplotlib.pyplot as plt
```

#### 1. Theorem 1.1: infinite line
Considering $J_n(x)$ as the n-th Bessel function of the first kind, we define
\begin{equation}
\tilde{J}_n = i^n J_n(z)
\end{equation}
and we have
\begin{equation}
\psi(x,t) = \tilde{J}_{|x-x_0|}(2t)
\end{equation}
so
$$
\psi(x,t) = i^{|x-x_0|}J_{|x-x_0|}(2t)
$$
by considering an adjacency matrix and initial condition $|x_0\rangle$.

#### 1. Theorem 1.1 (B): infinite line
Considering $J_n(x)$ as the n-th Bessel function of the first kind, we define
\begin{equation}
\tilde{J}_n = i^n J_n(z)
\end{equation}
and we have
\begin{equation}
\psi(x,t) = \tilde{J}_{|x-x_0|}(2t)
\end{equation}
so
$$
\psi(x,t) = \frac{i^{|x-x_0|}J_{|x-x_0|}(2t) + i^{|x-x_0-1|}J_{|x-x_0-1|}(2t)}{\sqrt{2}}
$$
by considering an adjacency matrix and initial condition 
$$\frac{1}{\sqrt{2}}(|x_0\rangle + |x_0 + 1\rangle$$.

#### 1. Theorem 1.1 (C): infinite line
Considering $J_n(x)$ as the n-th Bessel function of the first kind, we define
\begin{equation}
\tilde{J}_n = i^n J_n(z)
\end{equation}
and we have
\begin{equation}
\psi(x,t) = \tilde{J}_{|x-x_0|}(2t)
\end{equation}
so
$$
\psi(x,t) = \sum_{y \in L} \alpha_y i^{|x-y|}J_{|x-y|}(2t)
$$
by considering an adjacency matrix and initial condition 
$$|\psi_0\rangle = \sum_{y \in L}\alpha_y|y\rangle$$.

```python
def inf_line_bessel(x0,nx,t,theta):
    domain = np.arange(x0 - nx, x0 + nx + 1)
    psi = np.zeros([2*nx + 1,1], dtype='complex')
    
    for x in range(2*nx + 1):
        psi[x,0] = 1j**(np.abs(domain[x]-x0))*sp.jv(np.abs(domain[x] - x0), 2*t)/np.sqrt(2)
        psi[x,0] += 1j**(np.abs(domain[x]-x0-1))*sp.jv(np.abs(domain[x] - x0 - 1), 2*t)/np.sqrt(2)
        
        
    
    return np.abs(np.array(psi))**2
```

```python
x0 = 0
nx = 100
t  = 30
theta=np.pi/2

psi = inf_line_bessel(x0,nx,t,theta)
plt.plot(np.arange(x0 - nx, x0 + nx + 1), psi)
```
