```python
import numpy as np
import scipy.special as sp
import matplotlib.pyplot as plt
```

#### 1. Theorem 1.1: infinite line
Considering $J_n(x)$ as the n-th Bessel function of the first kind, we define
\begin{equation}
\tilde{J}_n = i^n J_n(z)
\end{equation}
and we have
\begin{equation}
\psi(x,t) = \tilde{J}_{|x-x_0|}(2t)
\end{equation}
so
$$
\psi(x,t) = i^{|x-x_0|}\tilde{J}_{|x-x_0|}(2t)
$$
by considering an adjacency matrix.

```python
def inf_line_bessel(x0,nx,t):
    domain = np.arange(x0 - nx, x0 + nx + 1)
    psi = np.zeros([2*nx + 1,1], dtype='complex')
    
    for x in range(2*nx + 1):
        psi[x,0] = 1j**(np.abs(domain[x]-x0))*sp.jv(np.abs(domain[x] - x0), 2*t)
    
    return np.abs(np.array(psi))**2
```

```python
x0 = 0
nx = 100
t  = 20
psi = inf_line_bessel(x0,nx,t)
plt.plot(np.arange(x0 - nx, x0 + nx + 1), psi)
```
